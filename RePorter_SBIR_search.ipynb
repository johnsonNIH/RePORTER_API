```python
#import packages needed
import requests
import json
from datetime import datetime
from datetime import date
from dateutil.relativedelta import relativedelta
```

```python
today = date.today()

#change the date in number of month in the relativedelta function
prev_date = date.today() + relativedelta(months=-8)
print("Today's Date :", today)
print("Previous Date :", prev_date)
```

```python
'''
Query the Reporter API

API documentation:
https://api.reporter.nih.gov/?urls.primaryName=V2.0
'''

url = "https://api.reporter.nih.gov/v2/projects/search"
headers = {'accept': 'application/json', 'Content-Type': 'application/json'}

#change the query here to extract the information you need, and assign to variable 'data'
data = {
     "criteria":
     {"fiscal_years": [2021, 2022],
         "funding_mechanism": [ "SB"],
         "agencies": [
         "NIAAA","NIDCD","NIDA","NIMHD","NLM","CIT","CSR","CLC","OD","NCCIH","NCMHD",
             "WH","FIC","NICHD","NCI","NCATS","NCCAM","NCRR","NEI","NHLBI","NHGRI", 
             "NIAID","NIAMS","NIBIB","NIDCR","NIDDK","NIEHS","NIGMS","NIMH","NINDS","NINR","NIA"
       ],
         "project_end_date": {
         "from_date": prev_date.strftime("%Y-%m-%d"),
         "to_date": today.strftime("%Y-%m-%d")
       },
         "include_active_projects": True,
      "is_agency_admin": True

     },
    "include_fields": [
        "ProjectTitle", "Organization", "ContactPiName","AgencyIcAdmin",
       "ProjectStartDate","ProjectEndDate"
     ],
     "offset":0,
     "limit":500,
     "sort_field":"project_end_date",
     "sort_order":"desc"
}
```

```python
#executing curl command with python codes, and transfer into json file
r = requests.post(url, data=json.dumps(data), headers=headers)
r_json = r.json()
```
